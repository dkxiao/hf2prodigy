```python
from datasets import load_dataset
import pandas as pd
```

```python
shao_data = load_dataset('limsc/requirements-entity-recognition')
```

```python
xiao_data = load_dataset('limsc/requirements-entity-recognition')
```

```python
# save df as pandas
df_shao_train = shao_data['train'].to_pandas()
df_shao_test = shao_data['test'].to_pandas()
df_shao_val = shao_data['val'].to_pandas()
```

```python
# function to transform df
def transform_df(df_input):
    df_output = df_input.drop(['id', 'ner_tags'], axis=1)
    df_output = df_input.rename(columns={"tokens": "text"})
    return df_output
```

```python
# function to detokenize function with regex
import re
from nltk.tokenize.treebank import TreebankWordDetokenizer as Detok

def detokenize(token_list):
    detokenizer = Detok()
    text = detokenizer.detokenize(token_list)
    text = re.sub('\s*,\s*', ', ', text)
    text = re.sub('\s*\.\s*', '.', text)
    return text
```

```python
# apply functions on all dataframes
df_shao_train = transform_df(df_shao_train)
df_shao_test = transform_df(df_shao_test)
df_shao_val = transform_df(df_shao_val)


df_xiao_train = df_shao_train['text'].apply(lambda x: detokenize(x))
df_xiao_test = df_shao_test['text'].apply(lambda x: detokenize(x))
df_xiao_val = df_shao_val['text'].apply(lambda x: detokenize(x))
```

```python
# display
with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.max_colwidth', 1000):
    print(df_xiao_train)
```

```python
# turn into json export file
df_xiao_train.to_json(r'./ner_req_train.json', orient='table', index=False)
df_xiao_test.to_json(r'./ner_req_test.json', orient='table', index=False)
df_xiao_val.to_json(r'./ner_req_val.json', orient='table', index=False)

# manual removal of schema necessary, including curly bracket {} at the end
```
